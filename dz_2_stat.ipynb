Написать **линейную регрессию** на два атрибута. Допустим, мы хотим предсказать цену дома (MEDV) в Бостоне, зная что цена на дом имеет линейную зависимость с атрибутом LSTAT. 



1.   Разделить данные на 80% тренировочных, 20% тестовых
2.   Составить модель линейной регрессии
3.   Посчитать ошибку модели на тестовых данных
4.   Сделать точечный график данных и полученной прямой линейной регрессии



```python
medv = data['MEDV']
lstat = data['LSTAT']
sns.regplot(x = lstat,y = medv,data = data, fit_reg = True)
plt.xlabel('LSTAT')
plt.ylabel('Цена дома')
plt.title('Зависимость цены дома от показателя LSTAT')
#plt.grid(False)
plt.show()
```

```python
medv = np.array(medv).reshape(-1,1)
lstat = np.array(lstat).reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
from sklearn.model_selection import train_test_split
```

Деление данных на 80% тренировочных и 20% тестовых

```python
X_train, X_test, y_train, y_test = train_test_split(
    lstat,
    medv,
    test_size = 0.2
)
```

Модель линейной регрессии

```python
model.fit(X_train, y_train)
```

```python
predicted_y = model.predict(X_test)
#predicted_y
```

Ошибка модели

```python
model.score(X_test, y_test)
```

График линейной регрессии

```python
plt.xlabel('lstat')
plt.ylabel('medv')
plt.title('Линейная регрессия от одной переменной')
x = np.linspace(-10, 60, 100)
y = model.coef_*x + model.intercept_
y = np.transpose(y)
plt.plot(x,y)
plt.plot(X_train, y_train, 'r.')
plt.plot(X_test, y_test, 'go')
plt.xlim(-10,50)
plt.ylim(-10,60)
plt.legend(['Прямая линейной регрессии', 'Тренировочные данные', 'Тестовые данные'])
plt.show()
```

# Дискретное распределение случайной величины

Напишите код для моделирования 10 подбрасываний игрального кубика. Начертите диаграму histogram. Какое это распределение?

Очень отдаленно напоминает геометрическое распределение

```python
import random
random.seed(1) # не поняла, что значит цифра после seed().

flips_till_heads = stats.geom.rvs(size=10,  # 10 подбрасываний кубика
                                  p=0.1667) # с вероятностью успеха 0.1667 = 1 / 6 - одна грань из шести

# Напечатать данные
print( pd.crosstab(index="counts", columns= flips_till_heads))

# Частотная зависимость
pd.DataFrame(flips_till_heads).hist(range=(-0.5,max(flips_till_heads)+0.5)
                                    , bins=max(flips_till_heads)+1)
plt.xlabel('Количество испытаний до первого успеха')
plt.ylabel('Частота встречаемости')
plt.title('Сгенерированное геометрическое распределение')
plt.show()
```

моделируйте 1000 раз сумму по 100 подбрасываний. Начертите диаграму histogram. Почему график изменился? Почему можно применить центральную предельную теорему?

Можно применить центральную предельную теорему, т.к.у нас есть большое количество слабо зависимых величин. Т.е.их можно просуммировать и получить нормальное распределение.

```python
# Функция для 100 подбрасываний
def geom_dev():
    random.seed(1)
    flips_till_heads = stats.geom.rvs(size=100,p=0.1667) 
    return flips_till_heads
geom_dev
```

```python
number_of_sums = 1000
X1 = np.zeros([number_of_sums])
for i in range(number_of_sums):
    X1[i] = np.sum(geom_dev())
plt.hist(X1,30)
plt.xlabel('X values')
plt.ylabel('Number of elements in each X')
plt.show()
```
