## Задание 2

Напишите функцию, возвращающую сумму первых n чисел Фибоначчи (для этого вам может понадобится ознакомиться с рекурсией).

```python
#рекурсивная функция по поиску числа Фибоначчи по его порядковому номеру х
def fib(x):
    if x == 1:
        return 0
    elif x == 2:
        return 1
    return fib(x-1) + fib(x-2)

#Ввод количества чисел Фибоначчи
n = int(input("Введите количество чисел Фибоначчи "))
#переменная х для цикла, для расчета чисел Фибоначчи для каждого ее значения
x = 1
#пустой список для наполнения его значениями
fib_line = []
#цикл для наполнения списка
while x <= n:
    fib_line.append(fib(x))
    x = x + 1
print("Первые ", n,"чисел Фибоначчи -" ,(fib_line))
```

## Задание 3


Напишите функцию, преобразующую произвольный список вида ['2018-01-01', 'yandex', 'cpc', 100] (он может быть любой длины) в словарь {'2018-01-01': {'yandex': {'cpc': 100}}}

```python
#функция для сбора массива данных из предложенного списка
def print_massive(data):
    data_len = len(data)
    # счетчик для цикла
    i = 1
    #определяю "нулевой" массв: т.е.это массив при минимуме данных (если есть только два значения в списке)
    list_mass = {data[data_len - (i + 1)]:data[data_len - (i)]}
    #цикл для проверки отличия длины списка от счетчика 
    while data_len - i >= 0 and (data_len -(i+2)) >= 0:
            #условие проверки минимальной длины
            if data_len == 2:
                list_mass = list_mass
                return list_mass
            #при невыполнении условия добавляю в сущестыующий массив новое значение, которое является массивом от предыдущей
            #итерации
            list_mass = {data[data_len - (i + 2)]:list_mass}
            i = i+1
    return list_mass
#Вызов функции
data = ['2018-01-01', 'yandex', 'cpc',100]
print_massive(data)
```
